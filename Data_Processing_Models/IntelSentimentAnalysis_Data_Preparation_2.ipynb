```python
review_list=i9['reviewDescription'].to_list()
len(review_list)
```

```python
languages=[]
from langid import classify

for i in review_list:
  for_classify=i
  language,_=classify(for_classify)
  languages.append(language)
```

```python
set(languages)
```

```python
i9['Langauge']=languages
```

```python
i9
```

```python
i9 = i9[i9["Langauge"]=='en']
```

```python
i9=i9.reset_index(drop=True)
i9.index = i9.index + 1
i9
```
