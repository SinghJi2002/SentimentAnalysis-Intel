```python
import pandas as pd
```

```python
df1=pd.read_csv("i3_12.csv")
df2=pd.read_csv("i3_13.csv")
df3=pd.read_csv("i3_14.csv")

df4=pd.read_csv("i5_12.csv")
df5=pd.read_csv("i5_13.csv")
df6=pd.read_csv("i5_14.csv")

df7=pd.read_csv("i7_12.csv",encoding='unicode_escape')
df8=pd.read_csv("i7_13.csv")
df9=pd.read_csv("i7_14.csv")

df10=pd.read_csv("i9_12.csv")
df11=pd.read_csv("i9_13.csv")
df12=pd.read_csv("i9_14.csv")
```

```python
intel_i7=pd.concat([df7,df8,df9],ignore_index=True)
intel_i5=pd.concat([df4,df5,df6],ignore_index=True)
intel_i9=pd.concat([df10,df11,df12],ignore_index=True)
intel_i3=pd.concat([df1,df2,df3],ignore_index=True)
```
